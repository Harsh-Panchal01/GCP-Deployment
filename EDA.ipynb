```python
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='Tenure', hue='Exited', palette='pastel')
plt.title('Churn comparison for different Tenure')
plt.xlabel('Tenure')
plt.ylabel('Exited counts')
plt.show()
```

The chart above indicates that individuals who chose to stay have a higher number of tenures compared to those who exited.

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='NumOfProducts', hue='Exited', palette='pastel')
plt.title('Churn comparison for different NumOfProducts')
plt.xlabel('NumOfProducts')
plt.ylabel('Exited counts')
plt.show()
```

The chart above illustrates that the highest rate of churn is observed among individuals with only one product. As the number of products increases, the rate of churning decreases. Notably, individuals who stayed as customers predominantly use two products. Interestingly, those who stayed are not using four products, and there is also a relatively small number of individuals who stayed using three products.

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='HasCrCard', hue='Exited', palette='pastel')
plt.title('Churn comparison for different credit cards')
plt.xlabel('HasCrCard')
plt.ylabel('Exited counts')
plt.show()
```

The churning rate is notably higher among credit card holders. This could be attributed to factors such as the inability to manage credit card interest or the possibility of finding more attractive offers from other sources.

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='IsActiveMember', hue='Exited', palette='pastel')
plt.title('Churn comparison for active/non active member')
plt.xlabel('IsActiveMember')
plt.ylabel('Exited counts')
plt.show()
```

Inactive customers have a higher tendency to leave the bank.

```python
df.groupby('Age')['Exited'].sum().to_frame().sort_values(by='Exited', ascending = False).head(20)
```

The highest number of churn cases can be observed within the age range of 35 to 56.

```python
df.groupby('Tenure')['Exited'].sum().to_frame().sort_values(by='Exited', ascending = False)
```

The highest number of churn cases is observed when the tenure is just 1 year. However, it's noteworthy that individuals are also leaving the bank even with a tenure of 9 years.

```python
df.groupby('NumOfProducts')['Exited'].sum().to_frame().sort_values(by='Exited', ascending = False)
```

When a customer uses a less number of products, the churn rate tends to be higher, and it decreases as the number of products increases.
